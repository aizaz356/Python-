```python
def fib(n):
    if n <= 1:                       # same as if n == 0 or n == 1:
        return n
    
    
    else:
        a = fib(n-1)
        b = fib(n-2)
        return a + b
        


v = fib(3)
```

```python
def fib2(n):
    if n<= 1: return n
 
    a,b = 0, 1
    
    for i in range(1, n+1):
        n = a + b
        a = b
        b = n
        
    return a
    
```
