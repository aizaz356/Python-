```python
seq = [1, 2, 8, 4, 6, 8, 9, 15, 'a']
```

```python
def count(s, value):
    """Count how many times value occurs in s"""
    counter = 0
    
    for i in s:
        
        if i == value :
            counter = counter + 1
            
    return counter
```

```python
count(seq, 8) 
```

```python
count(seq, 1)
```

```python
count(seq, 15)
```

```python
count(seq, 9)
```

```python
count(seq, 88)
```

```python
count(seq, 'a')
```

```python
count(seq, None)
```
