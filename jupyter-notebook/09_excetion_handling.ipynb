```python
BUBBLING EXCEPTIONS:
```

```python
def calculator(op, n1, n2):
    if op == 'div':
        return div(n1, n2)
    elif op == 'mul':
        return mul(n1, n2)
```

```python
print(calculator('div', 4, 2))
```

```python
print(calculator('div', 4, 0))
```

```python
try:
    print("Div 4 by 0 is:", calculator('div', 4, 0))
except ZeroDivisionError:
    print("Do not ask me to divide by 0 please.")
```

```python
x = 0
try: 
    print("Div 4 by 0 is:", calculator('div', 4, x))
except ZeroDivisionError: 
    print("Do not ask me to divide by 0 please.")
    
print("Done")
```

```python
# remember our square root example:
def sqrt(x, guess = 1.0):
    """Calculate square root of a positive number.
        Raises a ValueError if a negative number is passed to it."""
    
    if x < 0:
        raise ValueError("Cannot find square root of negative numbers")
    
    
    # calculate square eoot and return value
    return 6
```

```python
          # try
sqrt(-2)  # it's the caller's responsibility to handle the exception
```

```python
try:         # try
    sqrt(-2)  # it's the caller's responsibility to handle the exception
except ValueError:
    print("Tried to find square root of negative number")
```
