```python
def is_weekday(day, month, year):
    import datetime
    return datetime.datetime(year=year, month=month, day=day).weekday() < 5
```

```python
if is_weekday(28, 9, 2017) :
    print ("It's a weekday.")
else:
    print ("weekend!!")
```
